```python
from scipy.signal import filtfilt, butter
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fs = 32
n = 0.5*fs
f_high = 0.5
cut_off = f_high / n
order = 4
```

```python
def filtering(signal):
    b, a = butter(order, cut_off, btype='high')
    signal = filtfilt(b, a, signal)
    signal = (signal - np.min(signal)) / (np.max(signal) - np.min(signal))
    return signal

def helper(row):
    row = np.array(row)
    annotation = row[-1]
    row = row[:-1]
    row = np.reshape(row, (18, 256))
    return np.append(np.array([filtering(x) for x in row]), annotation)
```

```python
%%time 
df = pd.read_csv('babydf8sec.csv')
```

```python
df.drop(columns=['Unnamed: 0'],inplace=True)
```
